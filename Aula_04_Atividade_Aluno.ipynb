Crie um DataFrame que relacione os meses do ano com os gastos mensais. Após criado o DataFrame, apresente esses dados em pelo menos 3 tipos de gráficos. 

```
import pandas as pd
import matplotlib.pyplot as plt

dados = {"Mês": ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
         "Gastos (R$)": [4500, 4000, 3000, 3100, 3050, 3200, 3500, 3100, 3200, 3200, 3500, 4550]}

df = pd.DataFrame(dados, columns=['Mês', 'Gastos (R$)'])
df.plot.line(x='Mês', y='Gastos (R$)')
plt.show()

df.plot.bar(x='Mês', y='Gastos (R$)')
plt.show()

df.plot.scatter(x='Mês', y='Gastos (R$)')
plt.show()

```
